```python
import sys

def Input_Data():
	readl = sys.stdin.readline
	R, C = map(int,readl().split())
	map_city = [readl().strip() for _ in range(R)]
	return R, C, map_city


sol = -1
# 입력 받는 부분
R, C, map_city = Input_Data()

# 여기서부터 작성

from collections  import deque
direction = [(1,0),(-1,0),(0,1),(0,-1)]
point_dict = {} #지점들의 위치. dict를 이용한다. 
s_pos = None  #시작점 값


def do_bfs(sr,sc,R,C,map_city):
	cost_map = [ [-1]*C for _ in range(R) ]
	cost_map[sr][sc] = 0 ##시작위치 초기화
	distanceList = [0]*10	
	q = deque()	
	q.append((sr,sc))
	
	while q:
		r, c = q.popleft()
		for x,y in direction:			
			nr = r+x
			nc = c+y
			if 0 <= nr < R and 0<= nc < C and cost_map[nr][nc] == -1 and map_city[nr][nc] !='*':
				cost_map[nr][nc] = cost_map[r][c]+1				
				q.append((nr,nc))
				
	return cost_map
	

grid = [ [-1]*C for _ in range(R) ]
print(grid)
				
#크기와 맵을 입력받았다. 
#우선 맵을 전처리하자. 
for r in range(R):
	for c in range(C):
		ch = map_city[r][c]
		if ch == "S":
			s_pos = (r,c)
			digit = 0
			point_dict[digit] = (r,c)
			grid[r][c] = 0
		elif ch.isdigit():
			digit = int(ch)
			grid[r][c] = digit
			point_dict[digit] = (r,c)


relative_dist = [[10000]*len(point_dict) for _ in range(len(point_dict))]

#각 거리별로 다른지점간의 거리값을 계산하자. 
for i in range(len(point_dict)):
	sr, sc = point_dict[i]
	cost_map = do_bfs(sr,sc,R,C,map_city)
	
	for j in range(len(point_dict)):
		x, y = point_dict[j]
		relative_dist[i][j]=cost_map[x][y]
	

print(relative_dist)
	


# 출력하는 부분
print(sol)
```
