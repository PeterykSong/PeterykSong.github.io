```python
#1000~9999까지 약수의 개수를 구하라. 

# 약수의 의미 : 어떤수 N 이 i 에 의해 나누어졌을때, 나머지가 0 이면 약수다. 

q = []

for N in range(1000,9999):
    count = 0
    for i in range(1,N):
        if not (N % i):
            count += 1
    q.append([N,count])
```

```python
print(q)
```

```python
# 이 다음에는 행성간 탐색을 BFS를 통해 풀면 된다. 각 자리수를 아스키로 분해하는건 뭘까. 

#숫자를 아스키코드로 분해하는 건 파이썬에선 의외로 쉽다. 

a = 3256
b = 3456
sa = str(a)
sb = str(b)
count = 0
print(sa,sb)
print(list(sa))
print(list(zip(sa,sb)))
for x,y in zip(sa,sb):
    print(f'x {x},  y {y}')
    if x == y:
        count+=1
print(count)
```
