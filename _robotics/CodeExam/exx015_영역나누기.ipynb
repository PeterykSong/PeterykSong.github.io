```python
import sys

def InputData():
	readl = sys.stdin.readline
	N = int(readl())
	info = input()
	return N, info

sol = -1

# 입력
# N: 이동 정보의 개수
# info: 이동 방향 정보
N, info = InputData()

# 코드 작성

from collections import deque 


def bfs_zero(sx, sy, grid):
		q = deque([(sx, sy)])
		grid[sx][sy] = 3  # 방문 마킹(0→3)
		size = 1
		W, H = len(grid), len(grid[0])
		dirs = [(1,0), (-1,0), (0,1), (0,-1)]
		while q:
				x, y = q.popleft()
				for dx, dy in dirs:
						nx, ny = x + dx, y + dy
						if 0 <= nx < W and 0 <= ny < H and grid[nx][ny] == 0:
								grid[nx][ny] = 3
								size += 1
								q.append((nx, ny))
		return size


Direction = [[1,0],[-1,0],[0,-1],[0,1]]
route = []
info_list = list(info)
myX = 0
myY = 0
route.append([myX,myY])
for i in info_list:
	for j in range(3):
		myX = myX + Direction[int(i)][0]
		myY = myY + Direction[int(i)][1]
		route.append([myX,myY])

routeX = [p[0] for p in route]
routeY = [p[1] for p in route]
maxX,minX = max(routeX),min(routeX)
maxY,minY = max(routeY),min(routeY)

spanX = maxX - minX
spanY = maxY - minY
#offset 연산 
newRoute = [[x-minX+1, y-minY+1] for x,y in route]  #Map 작성을 위한 좌표값 offset (모두 양수로 이동)

grid = [[0 for _ in range(spanY+3)] for _ in range(spanX+3)]

for x,y in newRoute:
	grid[x][y] = 1

components = 0
for x in range(len(grid)):
	for y in range(len(grid[0])):
		if grid[x][y] == 0:
			components +=1
			bfs_zero(x,y,grid)
		
	
sol = components-1  #테두리 밖 공백영역은 1개다. 
# 출력
print(sol)
```
